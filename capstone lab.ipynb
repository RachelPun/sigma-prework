```python
opposite = wd.synsets("opposite")

print("synsets:\n", opposite)

opposite_lemmas = []
for synset in opposite:
    opposite_lemmas.append(synset.lemmas())
print("synsets' lemmas:\n", opposite_lemmas)

opposite_lemmas_names = []
for synset in opposite:
    opposite_lemmas_names.append(list(map(
        lambda lemma: lemma.name(),
        synset.lemmas()
    )))
print("synsets' lemmas' names:\n", opposite_lemmas_names)
print("synsets' lemmas' names:\n", )

print("\n")

print("synset2's name:\n", opposite[1].name())
# print("synset2's lemmas:\n", opposite[1].lemmas())
# print("synset2's lemmas' names:\n", list(map(
#     lambda lemma: lemma.name(),
#     opposite[1].lemmas())
# ))
print("synset2's hypos:\n", opposite[0].hyponyms())
print("synset2's hypos' lemmas:\n", list(map(
    lambda hyper: hyper.lemmas(),
    opposite[0].hyponyms()
)))
```

## similarity testing

```python
print("\narum.n.02", "arum.n.02")

print(wd.synset("arum.n.02").path_similarity(wd.synset("arum.n.02")))
print(wd.synset("arum.n.02").shortest_path_distance(wd.synset("arum.n.02")))

print("\narum.n.02", "catmint.n.01")

print(wd.synset("arum.n.02").path_similarity(wd.synset("catmint.n.01")))
print(wd.synset("arum.n.02").shortest_path_distance(wd.synset("catmint.n.01")))

print("\narum.n.02", "plantain.n.02")

print(wd.synset("arum.n.02").path_similarity(wd.synset("plantain.n.02")))
print(wd.synset("arum.n.02").shortest_path_distance(wd.synset("plantain.n.02")))

print("\narum 2", "granny_smith 2")

print(wd.synset("arum.n.02").path_similarity(wd.synset("granny_smith.n.01")))
print(wd.synset("arum.n.02").shortest_path_distance(wd.synset("granny_smith.n.01")))

print("\narum 2", "hello 1")

print(wd.synset("arum.n.02").path_similarity(wd.synset("hello.n.01")))
print(wd.synset("arum.n.02").shortest_path_distance(wd.synset("hello.n.01")))


print("\narum 1", "arum 2")

print(wd.synset("arum.n.01").path_similarity(wd.synset("arum.n.02")))
print(wd.synset("arum.n.01").shortest_path_distance(wd.synset("arum.n.02")))
```
